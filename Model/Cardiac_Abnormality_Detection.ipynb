# Cholestrol Level Prediction Using Machine Learning
To determine the cholestrol level normally a blood sample is taken and then it is determined through certain diagnostic procedures. But taking a blood sample is an invasive technique and is painful. So, to reduce the use of such invasive techniques this project aims to use Machine Learning to build an intellegent system that can predict the cholestrol levels based on age , gender , chest pain type , maximum resting heart rate. 

## Importing the Libraries

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
```

## Reading the dataset and making the feature and target vectors

```python
dataset = pd.read_csv("heart.csv")

feature1 = np.array(dataset["age"]).reshape(-1,1) # Age 
feature2 =  np.array(dataset["sex"]).reshape(-1,1) # Gender -> 1: Male , 0 : Female
feature3 =  np.array(dataset["cp"]).reshape(-1,1) # Chest Pain Type : 0-3 
feature4 =  np.array(dataset["trestbps"]).reshape(-1,1) #Max resting heart rate

x = np.concatenate((feature1,feature2,feature3,feature4),axis = 1)

y = np.array(dataset["chol"]).reshape(-1,1) # Max heart rate threshold
```

## Calling the regressor instance and fitting the data

```python
regressor = DecisionTreeRegressor(random_state=3)
regressor.fit(x,y)
```

## Evaluating the model

```python
regressor.score(x,y) #Model eval
```

## Visualizing the prediction

```python
import matplotlib.patches as mpatches

blue_patch = mpatches.Patch(color='blue', label='Ground Truth')
orange_patch = mpatches.Patch(color='orange', label='Prediction')

p1 = plt.plot(y)
p2 = plt.plot(regressor.predict(x))

plt.legend(handles=[blue_patch , orange_patch])

plt.show()
```

## Testing on collected data {Mom's Reports}

```python
testvec = np.concatenate((np.array(41).reshape(1,-1) , np.array(0).reshape(1,-1) , np.array(1).reshape(1,-1) , np.array(140).reshape(1,-1)) , axis = 1)
```

## Prediction

```python
regressor.predict(testvec) #Testing with real time data @MOMs reports that I put in the video
```
